## 2) Degradation Measurement of Robot Arm Position Accuracy

Create a unified dataset from all experimental files.

### 2a) Load all .csv files containing the raw data.

```python
path_raw_data = "data/robot_position_accuracy" # creat a path varriable called path_raw_data and set it to the path of the folder containing the raw data for the robot position accuracy project

path_raw_data = Path("data/robot_position_accuracy") # create a Path object from the path string using the Path() function from the pathlib library.

# Save all created dataframes from the csv-files in a dictionary with the name of the file as key and the dataframe as value
# we use a dictionary comprehension to create a dictionary called dfs 
# where the keys are the stem of the file names (i.e., the file name without the extension) 
# and the values are the dataframes created by reading each csv file in the path_raw_data directory using pandas read_csv function.
# The glob method is used to find all csv files in the directory.
dfs = {f.stem: pd.read_csv(f) for f in path_raw_data.glob("*.csv")}
print("\nDataframes created from csv files in robot_position_accuracy:")
# loop through the items in the dfs dictionary 
# and print the name of each dataframe (i.e., the key) and its shape 
for name, df in dfs.items():
    print("Gespeicherte csv-Datei:", name, "  Shape:", df.shape)


```

### 2b) Extract relevant metadata from the filenames and add them in separate columns:
- Speed (e.g., HalfSpeed, FullSpeed)
- Load in lbs
- Cold start condition
- Trial number

```python
import re
# loop through the items in the dfs dictionary and extract information from the file names using regular expressions (regex).
for name, df in dfs.items():
    filename = name
    # use the re.match() function to match the filename against a regex pattern that captures 
    # the relevant information (coldstart, speed, payload, trial number) and store the matched groups in variables.
    #each bracketed part of the regex pattern corresponds to a group that captures a specific piece of information from the filename.
    match = re.match(r"ur5testresult_(coldstart_)?(fullspeed|halfspeed)_payload([\d.]+)lb_(\d+)", filename)
    # if the filename matches the expected pattern, extract the information and add it as new columns in the dataframe.
    if match:
        coldstart = match.group(1) is not None # check if the first group (coldstart) is present in the filename and store the result as a boolean in the variable coldstart
        speed = match.group(2) # extract the second group (speed) from the filename and store it in the variable speed (fullspeed or halfspeed)
        payload = float(match.group(3)) # extract the third group (payload) from the filename, convert it to a float, and store it in the variable payload
        trial_number = int(match.group(4)) # extract the fourth group (trial number) from the filename, convert it to an integer, and store it in the variable trial_number
        # Add the extracted information as new columns in the dataframe
        df["ColdStart"] = coldstart
        df["Speed"] = speed
        df["Payload_lb"] = payload
        df["TrialNumber"] = trial_number
        print(f"\n--- '{filename}' ---")
        print(f"ColdStart={coldstart}, Speed={speed}, Payload_lb={payload}, TrialNumber={trial_number}")
        #print(df.head())
    else:
        print(f"Filename '{filename}' does not match expected pattern.")
df.head()
# as you can see, the new columns "ColdStart", "Speed", "Payload_lb", and "TrialNumber" have been added
# the other columns using the first row of the dataframe as header.
# this problem will be fixed in the next step 
```

### 2c) Add column names from the master data.
Ensure consistent and easy-to-handle column names.

```python
import re

# Load column names from master header file
header_df = pd.read_excel("data/robot_position_accuracy/ur5testresult_header.xlsx")
# save the column names from the header_df dataframe into a list called raw_col_names using the columns attribute and the tolist() method.
raw_col_names = header_df.columns.tolist()  # 73 column names

# Create clean, snake_case column names
def clean_col_name(col):
    # convert the column name to lowercase, remove parentheses, replace spaces and hyphens with underscores, and remove leading/trailing underscores.
    col = col.lower()
    col = col.replace("(", "").replace(")", "")
    col = re.sub(r"[\s\-]+", "_", col)
    col = re.sub(r"_+", "_", col)
    return col.strip("_")
# use a list comprehension to create a new list called clean_col_names by applying the clean_col_name function to each column name in the raw_col_names list.
clean_col_names = [clean_col_name(c) for c in raw_col_names]
print("Clean column names:")
#iterate through the raw_col_names and clean_col_names lists simultaneously using the zip() function and print each pair of raw and clean column names in a formatted way.

for raw, clean in zip(raw_col_names, clean_col_names):
    print(f"  {raw!r:50s} -> {clean!r}")

df.head()
```

```python


# Reload all CSVs with correct column names (header=None, first row was data not header)
path_raw_data = Path("data/robot_position_accuracy")
# dictionary comprehension to create a dictionary called dfs 
# we load all csv files so that the first row is not treated as header but as data and we assign the clean column names from the header file-list(clean_col_names)
dfs = {f.stem: pd.read_csv(f, header=None, names=clean_col_names) for f in path_raw_data.glob("*.csv")} 

# we loop through all dataframes in the dictionary and convert all columns with object dtype to numeric, removing any non-numeric characters
for df in dfs.values(): 
    # we loop through all columns in the dataframe that have the object data type using the select_dtypes() method and the include="object" argument.
    # for each column that has the object data type, we use the str.replace() method to remove any non-numeric characters (except for digits, decimal points, and hyphens) from the column values using a regular expression.
    # then we use the pd.to_numeric() function to convert the cleaned column values to numeric
    # example: if a column has values like "1.23222]]", the str.replace() method will remove the " ]]" part and leave "1.23", which can then be converted to a numeric value using pd.to_numeric().
    for col in df.select_dtypes(include="object").columns:
        df[col] = pd.to_numeric(
            df[col].str.replace(r"[^\d.\-]", "", regex=True),
            errors="coerce"
        )

# Re-apply metadata from filenames (2b)
# this step is necessary because we reloaded the csv files with header=None, so we need to re-extract the metadata from the filenames and add it as new columns in the dataframes.
for name, df in dfs.items():
    match = re.match(r"ur5testresult_(coldstart_)?(fullspeed|halfspeed)_payload([\d.]+)lb_(\d+)", name)
    if match:
        df["cold_start"]   = match.group(1) is not None
        df["speed"]        = match.group(2)
        df["payload_lb"]   = float(match.group(3))
        df["trial_number"] = int(match.group(4))

print(f"\nReloaded {len(dfs)} DataFrames with {len(clean_col_names)} named columns + 4 metadata columns.")
print("Example shape:", next(iter(dfs.values())).shape)
print("Example columns:", next(iter(dfs.values())).columns.tolist())
df.head()

```
